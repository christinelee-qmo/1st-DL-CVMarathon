### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

## CSV
with open("Data/d12.csv", "w", newline="") as csvfile:
    for d in soup.find(class_="part_list_2").find_all('h3'):
        date = d.find(class_="date").text
        tag = d.em.text
        title = d.a.text
        wr = csv.writer(csvfile)
        wr.writerow([date, tag, title])

```

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

## JSON

res = {} #WHY?
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    tag = d.em.text
    title = d.a.text
    
    res.setdefault(tag, [])
    res[tag].append({
        'title': title,
        'date': date
    })
    
ress = print(res)

text_file = open("Data/d12.json", "w")
text_file.write(ress)
text_file.close()
```

### ③ 哪一個類別的文章最多

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

## JSON
res = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    tag = d.em.text
    title = d.a.text
    
    res.setdefault(tag, [])
    res[tag].append({
        'title': title,
        'date': date
    })
# print(res)



data = []
for i in res:
#     print(i,len(res[i])) # count, get individual counts
    d = {}
    d['tag'] = i
    d['count'] = len(res[i])
    data.append(d)

#列印出排序過的第一個項目
d_1 = sorted(data, key=lambda d: d['count'])[::-1][0]
print(d_1)
print(d_1['tag'])


# for d in sorted(data, key=lambda d: d['count'])[::-1]:
#     print(d)
```
