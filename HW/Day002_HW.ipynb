## CSV Read

```python
import csv
with open('./data/BusTime.csv', newline='') as csvfile: # 開啟 CSV 檔案
    rows = csv.reader(csvfile)     # 讀取 CSV 檔案內容
    for row in rows:    # 以迴圈輸出每一列
        print(row)
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存
import csv

with open('./data/BusTime.csv', newline='', encoding="utf-8") as csvfile:
    schedule = []
    rows = csv.reader(csvfile)
    next(rows)
    for row in rows: #在rows中循環所有的項目，命名row
        schedule.append(row[5])  #列表末尾添加新的對象 (why filter?)
    print(schedule)
```

# 3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

import csv

with open('./data/BusTime.csv', newline='', encoding="utf-8") as csvfile:
    schedule1 = []
    schedule2 = []
    
    rows = csv.reader(csvfile)
#     next(rows)
    for row in rows: #在rows中循環所有的項目，命名row
        schedule1.append(row[5])
        schedule2.append(row[6])
    print(schedule1)
    print(schedule2)
```

```python
#Json

import csv

with open('./data/BusTime.csv', newline='', encoding="utf-8") as csvfile:
    schedule = []
    rows = csv.reader(csvfile)
    for index, row in enumerate(rows):
        for i2, val in enumerate(row[5:-1]):
            if index == 0:
                schedule.append({ 'bus': val, 'times': [] })
            else:
                schedule[i2]['times'].append(val)
    print(schedule)
```
